```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv(r"C:\Users\khushi pawar\Downloads\concrete_data.csv")
```

```python
df.sample(3)
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(df[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(df[col],dist="norm",plot=plt)
    plt.grid(True)
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PowerTransformer
```

```python
X=df.drop(columns=['Strength'])
Y=df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
pt=PowerTransformer(method='box-cox')

X_train_trf=pt.fit_transform(X_train+0.00001)
X_test_trf=pt.transform(X_test+0.00001)

pd.DataFrame({'cols':X_train.columns,'box-cox':pt.lambdas_})
```

**kitna sahi predict kiya hai /ya model kaisa chal rha hai**

```python
clf=LinearRegression()

clf.fit(X_train_trf,y_train)

pred=clf.predict(X_test_trf)

r2_score(y_test,pred)
```

**ye cross validation ke liye bas use kra hai isko khi graph me use nhi krege**

```python
pt=PowerTransformer(method='box-cox')
X_trans=pt.fit_transform(X+0.00001)

clf=LinearRegression()
np.mean(cross_val_score(clf,X_trans,Y,scoring="r2"))
```

```python
X_train_trf=pd.DataFrame(X_train_trf,columns=X_train.columns)

for col in X_train_trf.columns:
    plt.figure(figsize=(14,4))
    #plt.subplot(121)
    #sns.kdeplot(X_train[col])
    #plt.title(col)

    plt.subplot(121)
    sns.kdeplot(X_train_trf[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(X_train_trf[col],dist="norm",plot=plt)
    plt.title(col)

    plt.show()
    
```

```python
pt=PowerTransformer()

pt_train=pt.fit_transform(X_train,y_train)
pt_test=pt.transform(X_test)

pd.DataFrame({'col':X_train.columns,'yeo-johnson':pt.lambdas_})
```

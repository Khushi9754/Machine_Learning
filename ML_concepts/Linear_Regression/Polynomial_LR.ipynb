```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

# Generate X values
X = 6 * np.random.rand(200, 1) - 3  # range [-3, 3]

# Generate cubic relationship: y = 0.5 * X^3 + 0.8 * X^2 + 0.9 * X + 2 + noise
y =  0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200, 1)
```

```python
X.shape
```

```python
y.shape
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=32)
```

```python
li = LinearRegression()
li.fit(X_train,y_train)

y_pred = li.predict(X_test)

print(" r2_score : ",r2_score(y_test,y_pred))
```

```python
plt.scatter(X,y)
plt.plot(X_train,li.predict(X_train),color='red')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

Polynomial wala

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
#best degree for this
pf = PolynomialFeatures(degree=2,include_bias=True)

#overfitting
#pf = PolynomialFeatures(degree=50,include_bias=True)

#underfitting
#pf = PolynomialFeatures(degree=1,include_bias=True)

X_train_trf = pf.fit_transform(X_train)
X_test_trf = pf.transform(X_test)
```

```python
print(X_train_trf[0])
print(X_train[0])
```

```python
lr = LinearRegression()

lr.fit(X_train_trf,y_train)
```

```python
y_pred1 = lr.predict(X_test_trf)

print("r2_score : ",r2_score(y_test,y_pred1))
```

```python
print("coef : ",lr.coef_)
print("intercept : ",lr.intercept_)

#y =  0.8 * X^2 + 0.9 * X + 2 + noise
```

```python
np.random.seed(42)

# Generate X values
X1 = 6 * np.random.rand(200, 1) - 3  # range [-3, 3]

# Generate cubic relationship: y = 0.5 * X^3 + 0.8 * X^2 + 0.9 * X + 2 + noise
y1 = 0.5 * X**3 + 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200, 1)
```

```python
X_train1,X_test1,y_train1,y_test1 = train_test_split(X1,y1,test_size=0.2,random_state=34)
```

```python
lir = LinearRegression()
lir.fit(X_train1,y_train1)

y_pred2 = lir.predict(X_test1)

print("r2_score : ",r2_score(y_test1,y_pred2))
```

```python
pl = PolynomialFeatures(degree=3,include_bias=False)

X_train_trf1 = pl.fit_transform(X_train1)
X_test_trf1 = pl.transform(X_test1)
```

```python
lir1 = LinearRegression()
lir1.fit(X_train_trf1,y_train1)

y_pred3 = lir1.predict(X_test_trf1)

print("r2_score : ",r2_score(y_test1,y_pred3))
```

```python
#y = 0.5 * X^3 + 0.8 * X^2 + 0.9 * X + 2 + noise
#with include_bias = True

print("coef : ",lir1.coef_)
print("intercept_ : ",lir1.intercept_)
```

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\khushi pawar\Downloads\train.csv")
```

```python
df.sample(5)
```

```python
df.dropna(inplace=True)
```

```python
X = df.drop(columns=['Survived','PassengerId','Name','Sex','Ticket','Fare','Cabin','Embarked'])
y= df['Survived']
```

```python
X
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

**feature consatruction**

```python
X['Family']= X['SibSp'] + X['Parch'] +1
```

```python
def cal(num):
    if (num==1):
        return 0
    elif (num>1) & (num<=4):
        return 1
    else:
        return 2
```

```python
X['Family_size']=X['Family'].apply(cal)
```

```python
X.drop(columns=['Family','SibSp','Parch'],inplace=True)
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

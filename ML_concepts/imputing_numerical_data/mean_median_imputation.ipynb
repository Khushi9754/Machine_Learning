```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv(r"C:\Users\khushi pawar\Downloads\titanic_toy.csv")
```

```python
df.sample(4)
```

```python
df.shape
```

```python
X=df.iloc[:,0:3]
Y=df.iloc[:,-1]
```

```python
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
trf=ColumnTransformer([
    ('mean_age',SimpleImputer(),['Age']),
    ('median_fare',SimpleImputer(strategy='median'),['Fare'])
],remainder='passthrough')
```

```python
X_trained=trf.fit_transform(X_train)
X_tested=trf.transform(X_test)
```

```python
combined_array = np.vstack((X_trained, X_tested))

new_df=pd.DataFrame(combined_array,columns=['Age','Fare','Family'])
```

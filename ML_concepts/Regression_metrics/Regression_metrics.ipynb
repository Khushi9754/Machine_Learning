```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"C:\Users\khushi pawar\Downloads\placement (1).csv")
```

```python
df.sample(5)
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('cgpa')
plt.ylabel('package')
```

```python
X = df.iloc[:,0:1]
Y = df.iloc[:,1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
y_predict = li.predict(X_test)
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,li.predict(X_train),color='red')
plt.xlabel('cgpa')
plt.ylabel('package')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
```

```python
print("MAE : ",mean_absolute_error(y_test,y_predict))
```

```python
print("MSE : ",mean_squared_error(y_test,y_predict))
print("RMSE : ",root_mean_squared_error(y_test,y_predict))
print("ME RMSE : ",np.sqrt(mean_squared_error(y_test,y_predict)))
```

```python
print("R2 Score : ",r2_score(y_test,y_predict))
r2 = r2_score(y_test,y_predict)
```

```python
# Adjusted R2 Score
adj = 1 - ( ((1-r2)*(X_test.shape[0]-1)) / (X_test.shape[0]-1-X_test.shape[1])  )
```
